```python
#import libraries 
import requests 
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#the .get function retireves the url with the table that needs to be scraped and then beautiful soup parses it for html   
url = 'https://www.espn.com/nfl/stats/team/_/view/offense/stat/passing/table/passing'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#the find all functions find all the tables with the 'tbody' tags with the calss 'Table_TBODY' and stores them in the list assigned to the variable all_tables.
#the first item in the list contains the names and the second item contains the all the data  
all_tables = soup.find_all('tbody', class_='Table__TBODY')
names = all_tables[0]
values = all_tables[1]
```

```python
#The empty lists will be appended with dae data form the tables that will then be used to create the dataframe 
teams = []
gp = []
completions = [] 
attempts = []
completion_percent = [] 
yards = [] 
average = [] 
yards_per_game = [] 
longest_pass = []
touchdowns = []
interceptions = [] 
sacks = []
sack_yards_lost = []
passer_rating= []

#the find all function finda all the rows in the names table then the for loop finds the values in each row an appends them to the 'teams' list  
names_trs = names.find_all('tr')
for x in names:
    teams.append(x.get_text())

#the first find all function finds all the rows in the values table then the for loop finds all the values in these rose then appends them to the approriate list  
values_trs = values.find_all('tr')
for x in values_trs:
  tds = x.find_all('td')
  gp.append(float(tds[0].get_text()))
  completions.append(float(tds[1].get_text()))
  attempts.append(float(tds[2].get_text()))
  completion_percent.append(float(tds[3].get_text()))
  yards.append(float(tds[4].get_text().replace(',', '')))
  average.append(float(tds[5].get_text()))
  yards_per_game.append(float(tds[6].get_text()))
  longest_pass.append(float(tds[7].get_text()))
  touchdowns.append(float(tds[8].get_text()))
  interceptions.append(float(tds[9].get_text()))
  sacks.append(float(tds[10].get_text()))
  sack_yards_lost.append(float(tds[11].get_text()))
  passer_rating.append(float(tds[12].get_text()))
```

```python
#creating a series that will used to create the dataframe 
passing_data = {
            'Team Name' : teams,
            'Games Played' : gp,
            'Completions' : completions,
            'Pass Attempts' : attempts,
            'Completion Percentage' : completion_percent,
            'Total Yards' : yards,
            'Average' : average,
            'Yards per Game' : yards_per_game,
            'Longest Pass' : longest_pass,
            'Touchdowns' : touchdowns,
            'Interceptions' : interceptions,
            'Sacks' : sacks,
            'Yards Lost' : sack_yards_lost,
            'Passer Rating' : passer_rating,
            }


#creates dataframe using pandas 
passing_df = pd.DataFrame(passing_data)
passing_df.to_csv ('PassingData.csv')
passing_df
```
