```python
#We will be using pandas for loading the csv file into a dataframe
import pandas as pd

#matplotlib is a python library for basic graph plotting
import matplotlib.pyplot as plt

#seaborn is another python graphing library that is built upon matplotlib and allows you to optimize matplotlib's output
import seaborn as sns 

#switching from matplotlib's default graphing style to that of seaborn's. Seaborn displays graph lines, matplotlib doesn't
sns.set(color_codes = True)

#removing warnings from the output
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"C:\Users\ramsa\OneDrive\Desktop\Spring 2023\CIS 3120\Project #2 Part B\StudentsPerformance.csv")
df
```

```python
#Graph 1
#relationship between test preparation courses and math scores 
sns.catplot(data=df, x="test preparation course", y="math score", order=["completed", "none"])
```

```python
#Graph 2 
#distribution of test scores between kids that have standard lunch vs free/reduced lunch
sns.displot(df, x="reading score", hue="lunch", kind="kde")
```

```python
#Graph 3
#Relationship betewen reading and writing scores 
sns.lmplot(x="reading score", y="writing score", data=df, aspect=3)
```

```python
#Graph 4 
#Relationship betwwen math and writing scores 
sns.lmplot(x="math score", y="writing score", data=df, aspect=3)
```

```python
#Graph 5
#difference of reading scores between male and female
fig, ax = plt.subplots(figsize=(6,9))
sns.boxplot(x='gender', y='reading score', palette='rocket', data=df)
```

```python
#Graph 6
#difference of writing scores between male and female
fig, ax = plt.subplots(figsize=(6,9))
sns.boxplot(x='gender', y='writing score', palette='rocket', data=df)
```

```python
#Graph 7
#How math scores vary between genders 
fig, ax = plt.subplots(figsize=(6,9))
sns.boxplot(x='gender', y='math score', palette='rocket', data=df)
```

```python
#8
#average of test scores of the students that completed a test prep course vs students that didnt
sns.catplot(data=df, x="test preparation course", y="writing score", kind="bar")
```

```python
#Graph 9
#distrubutions of students' parents education level
sns.catplot(data=df, x="parental level of education", kind="count", palette="ch:.25", aspect= 2)
```

```python
#Graph 10 
#distributions of scores among the different racial groups 
sns.displot(df, x="reading score", hue="race/ethnicity", kind="kde", fill=True)
```

```python
#Graph 11
#the relation ship between scores and lunch
fig, ax = plt.subplots(figsize=(6,9))
sns.boxplot(x='lunch', y='reading score', palette='rocket', data=df)
```

```python
#graph 12 
#how parents education affects the sudents scores 
sns.catplot(
    data=df, x="parental level of education", y="math score", hue="gender",
    kind="violin", split=True, aspect= 2)
```
